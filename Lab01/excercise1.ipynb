Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import data and set first column as the index

```python
df = pd.read_csv("Data1.csv", index_col=0)
df.head()
```

Plot all columns as time series

```python
df.plot(grid=True, figsize=(20,8), subplots=True, layout=(3,2), sharex=False)
plt.show()
```

Plot histograms of all colums

```python
df.hist(bins=20, figsize=(15,10))
plt.show()
```

Plot KDE for all colums

```python
df.plot.kde(subplots=True, grid=True, layout=(3,2), figsize=(15,10))
plt.show()
```

Repeat analysis for theta1-theta4 in 2018
- time series
- histogram
- kde

```python
df_2018 = df.loc['2018-01-01':'2018-12-31', "theta_1":"theta_4"]
df_2018.plot(grid=True, figsize=(20,6), subplots=True, layout=(2,2), sharex=False)
plt.show()
```

```python
df_2018.hist(bins=30, figsize=(10,10))
plt.show()
```

```python
df_2018.plot.kde(subplots=True, layout=(2,2), figsize=(10,10), grid=True, sharex=False)
plt.show()
```
